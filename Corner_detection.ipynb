```python
real_chess=cv2.imread('/content/drive/MyDrive/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/real_chessboard.jpg')
real_chess=cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)

flat_chess=cv2.imread('/content/drive/MyDrive/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/flat_chessboard.png')
flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

gray_flat_chess=cv2.cvtColor(flat_chess,cv2.COLOR_RGB2GRAY)
gray_real_chess=cv2.cvtColor(real_chess,cv2.COLOR_RGB2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,70,0.01,10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
  x,y = i.ravel()
  cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners=cv2.goodFeaturesToTrack(gray_real_chess,70,0.01,10)
corners=np.int0(corners)
```

```python
for i in corners:
  x,y=i.ravel()
  cv2.circle(real_chess,(x,y),3,(255,0,0),1)
```

```python
plt.imshow(real_chess)
```
