```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img,cmap=None):
  fig=plt.figure(figsize=(12,10))
  ax=fig.add_subplot(111)
  ax.imshow(img,cmap)
```
