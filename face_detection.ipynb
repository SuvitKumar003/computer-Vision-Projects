```python
face_Cascade=cv2.CascadeClassifier('DATA\haarcascades\haarcascade_frontalface_default.xml')
```

```python
import cv2

def detect_img1(img):
    # Check if the image is loaded properly
    if img is None:
        raise ValueError("Image not loaded properly. Check the file path or input image.")
    
    face_img = img.copy()
    
    # Detect faces
    face_rects = face_Cascade.detectMultiScale(face_img)
    
    # Draw rectangles around detected faces
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 255, 255), 10)
    
    return face_img

# Example of loading the image and calling the function
 # Ensure the path is correct
```

```python
denis1=cv2.imread("DATA/Nadia_Murad.jpg")
res = detect_img1(denis1)
```

```python
plt.imshow(res,cmap='gray')
```

```python
import cv2

def adjusted1_detect_img1(img):
    # Check if the image is loaded properly
    if img is None:
        raise ValueError("Image not loaded properly. Check the file path or input image.")
    
    face_img = img.copy()
    
    # Detect faces
    face_rects = face_Cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    # Draw rectangles around detected faces
    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 255, 255), 10)
    
    return face_img

# Example of loading the image and calling the function
 # Ensure the path is correct
```

```python
denis=cv2.imread("DATA/solvay_conference.jpg")
result=adjusted1_detect_img1(denis)
```

```python
plt.imshow(result)
```

```python
eye_cascade=cv2.CascadeClassifier('DATA\haarcascades\haarcascade_eye.xml')
```

```python
import cv2

def detect_eyes(img):
    # Check if the image is loaded properly
    if img is None:
        raise ValueError("Image not loaded properly. Check the file path or input image.")
    
    face_img = img.copy()
    
    # Detect faces
    eyes_rects = eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    # Draw rectangles around detected faces
    for (x, y, w, h) in eyes_rects:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 255, 255), 10)
    
    return face_img

# Example of loading the image and calling the function
 # Ensure the path is correct
```

```python
resutlt=detect_eyes(nadia)
```

```python
plt.imshow(resutlt,cmap='gray')
```

```python
cap=cv2.VideoCapture(0)
while True:
   ret,frame = cap.read(0)

   frame=adjusted1_detect_img1(frame)

   cv2.imshow(" Video frame detected",frame)

   k=cv2.waitKey(1)

   if k==27:
      break
   
cap.release()
cv2.destroyAllWindows()

```
