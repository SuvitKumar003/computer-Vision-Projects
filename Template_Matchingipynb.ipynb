```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
  full_copy=full.copy()
  method=eval(m)
  #template matching
  res=cv2.matchTemplate(full_copy,face,method)

  min_Val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
# the basic prupose for using the if else loop is that the functions mentioned in the conditon return the best result in the form of the min value but the others return the best result in the form of the max value so the assign the corner of the rectangle in the if else loop according to the best value returned by the loop
  if method in [cv2.TM_SQDIFF,cv2.TM_SQDIFF_NORMED]:
    top_left = min_loc
  else:
    top_left=max_loc

  height,width,channels=face.shape
  bottom_right=(top_left[0]+width,top_left[1]+height)

  cv2.rectangle(full_copy,top_left,bottom_right,(255,0,0),thickness=10)

  plt.subplot(121)
  plt.imshow(res)
  plt.title(" heatmap of the full picutre")
  plt.subplot(122)
  plt.imshow(full_copy)
  plt.title('detection of template')
  plt.suptitle(m)

  plt.show()
  print('\n')
```
